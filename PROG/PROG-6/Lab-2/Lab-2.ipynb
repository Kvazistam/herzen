### Статистика таблицы.


```python

df = pd.read_csv('train.csv')
df.describe()
```

```python
n = len(df['PassengerId'].unique())
```

### 1. Какое количество мужчин и женщин ехало на параходе? Приведите два числа через пробел.


```python
a = df.groupby("Sex")['PassengerId'].count()
mens = a['male']
womens = a['female']
# df[df['Sex'] == 'male']
print(f'Мужчин - {mens}',f'Женщин - {womens}')
```

### 2. Подсчитайте сколько пассажиров загрузилось на борт в различных портах? Приведите три числа через пробел.

```python
a = df.groupby("Embarked")['PassengerId'].count()

c = a['C']
q = a["Q"]
s = a['S']
print(c,q,s)
```

### 3. Посчитайте долю (процент) погибших на параходе (число и процент)?



```python
a = df.groupby("Survived")['PassengerId'].count()
died = a[0]
percent = died/n
print(f'Погибло {died}, процент - {round(percent, 3)*100}%')
```

### 4. Какие доли составляли пассажиры первого, второго, третьего класса?

```python
a = df.groupby("Pclass")['PassengerId'].count()
class_1 = round(a[1]/n, 3)
class_2= round(a[2]/n, 3)
class_3 = round(a[3]/n, 3)
print(f'Первый класс {class_1}, Второй класс {class_2}, Третий класс {class_3},')
```

### 5. Вычислите коэффициент корреляции Пирсона между количеством супругов (SibSp) и количеством детей (Parch).

```python
X = df['SibSp']
Y = df['Parch']

correlation = X.corr(Y, method="pearson")
correlation
```

### 6. Выясните есть ли корреляция (вычислите коэффициент корреляции Пирсона) между:
+ возрастом и параметром Survived;
+ полом человека и параметром Survived;
+ классом, в котором пассажир ехал, и параметром Survived.

```python
# Корреляция между возрастом и параметром Survived

X = df['Age']
Y = df['Survived']

correlation = X.corr(Y, method="pearson")
correlation
```

```python
# Корреляция между полом человека и параметром Survived

X = df['Sex'].map({'male': 1, 'female':0})
Y = df['Survived']

correlation = X.corr(Y, method="pearson")
correlation
```

```python
# Корреляция между классом, в котором пассажир ехал, и параметром Survived.

X = df['Pclass']
Y = df['Survived']

correlation = X.corr(Y, method="pearson")
correlation
```

### 7. Посчитайте средний возраст пассажиров и медиану, минимальный и максимальный возраст.

```python
a = df['Age']
print(f'Средний возраст: {a.mean():.3f}')
print(f'Медианный возраст: {a.median()}')
print(f'Минимальный возраст: {a.min()}')
print(f'Максимальный возраст: {a.max()}')
```

### 9. Какое самое популярное мужское имя на корабле?

```python

reg = r'\. \(*(\w*)\)*' # всегда полезно поупражняться в регулярных выражениях
mens = df[df['Sex']=='male']
# womens = df[df['Sex']=='female']
a = pd.DataFrame()
a['Name'] = mens['Name'].str.extract(reg)

Popular_Name, count = a.groupby('Name')['Name'].value_counts().idxmax(), a.groupby('Name')['Name'].value_counts().max()
print(Popular_Name, count)
```

### 10. Какие самые популярные мужское и женские имена людей, старше 15 лет на корабле?

```python
older_15 = df[df["Age"]>15]
reg = r'\. (\w+)(?![\w, ]*\()|\((\w+)'

mens = older_15[older_15['Sex']=='male']
womens = older_15[older_15['Sex']=='female']
import re
women_names = pd.DataFrame()
a = womens['Name'].str.extract(reg)
women_names['Name'] = a[0].fillna(a[1])
popular_women_name, count_women_name = women_names.groupby('Name')['Name'].value_counts().idxmax(), women_names.groupby('Name')['Name'].value_counts().max()

men_names = pd.DataFrame()
a = mens['Name'].str.extract(reg)
men_names['Name'] = a[0].fillna(a[1])
popular_men_name, count_men_name = men_names.groupby('Name')['Name'].value_counts().idxmax(), men_names.groupby('Name')['Name'].value_counts().max()

print(f'Самое популярное мужское имя у людей старше 15 лет - {popular_men_name} оно встречается {count_men_name} раз')
print(f'Самое популярное женское имя у людей старше 15 лет - {popular_women_name} оно встречается {count_women_name} раз')
```
