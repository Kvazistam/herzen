![Пример графика](https://community-cdn.rstudio.com/uploads/default/original/2X/5/5aa52f9ea780bec548919555e53787dc7ffac41b.png "Пример графика")

График должен сопровождаться легендой с объяснением каким цветом закодированы онлайн и оффлайн-продажи по вертикали должны быть обозначены текстом месяцы, а по горизонтали значения с соответствующим шагом (выберите подходящий шаг). Для визуализации данные за каждый день должны быть предварительно суммированы.

На графике должны присутствовать подписи для осей, метки с соответствующим шагом (выберите его самостоятельно), цвет точек и остальные параметры - по усмотрению студента. Подписи для меток для оси x можно обозначить как день года (от 1 до 365).

### График для _MarketingSpend.csv_

```python
import numpy as np
offline =  np.array(dfMarketing['Offline'])
online = np.array(dfMarketing['Online'])
dfMarketing["Date"] = pd.to_datetime(dfMarketing["Date"])
dfMarketing["Month"] = dfMarketing["Date"].dt.to_period("M")
monthly_sales = dfMarketing.groupby('Month')
offline_sales = np.array(monthly_sales['Offline'].sum())
online_sales = np.array(monthly_sales['Online'].sum())
months = dfMarketing['Month'].unique().astype(str)
```

```python
import matplotlib.pyplot as plt

x = np.arange(len(months))
width = 0.5

fig, ax = plt.subplots(figsize=(15, 6))
bars_offline = ax.bar(x - width/2, offline_sales, width, label='Offline')
bars_online = ax.bar(x + width/2, online_sales, width, label='Online')

for bar in bars_online:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
            f"{int(bar.get_height())}", ha="center", va="bottom")

for bar in bars_offline:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
            f"{int(bar.get_height())}", ha="center", va="bottom")

ax.set_xlabel("Дата")
ax.set_ylabel("Сумма продаж")
ax.set_title("Сумма онлайн и офлайн продаж по месяцам")
ax.set_xticks(x)
ax.set_xticklabels(months, ha='center')
ax.legend()

plt.tight_layout()
plt.show()
```
